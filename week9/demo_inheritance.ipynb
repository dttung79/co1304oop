```python
class A:
    def __init__(self):
        print('A constructor called')
        self.value = 'A'

    def display(self):
        print(f'My value: {self.value}')

# B inherits from A
class B(A):
    def __init__(self):
        super().__init__()  # Call the constructor of A
        print('B constructor called')
```

```python
a = A()  # Create an instance of A
a.display()  # Call display method of A
```

```python
b = B()  # Create an instance of B
```

```python
b.display()  # Call display method inherited from A
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def hi(self):
        print(f'Hi, I am {self.name}, and I am {self.age} years old.')
```

```python
john = Person('John', 30)
john.hi()
```

```python
class Student(Person):
    def __init__(self, name, age, grade):
        # call the constructor of the base (super) class to initialize name and age
        super().__init__(name, age)
        # set the additional attribute for Student
        self.grade = grade
    
    def hi(self):
        # call the hi method of the base class
        super().hi()
        # add additional information for Student
        print(f'My grade is {self.grade}.')
```

```python
s = Student('Alice', 20, 8.5)
s.hi()
```

```python
class A:
    def hello(self):
        print('Hello')
    
    def _greating(self):
        print('Greetings!')

    def __secret(self):
        print('This is a secret method.')
```

```python
a = A()  # Create an instance of A
a.hello()  # Call public method
```

```python
a._greating()  # Call protected method
```

```python
a.__secret()  # Attempt to call private method (will raise AttributeError)
```

```python
class B(A):
    def my_method(self):
        self.hello()  # Call inherited public method
        self._greating()  # Call inherited protected method
        self.__secret()  # Attempt to call inherited private method (will raise AttributeError)
```

```python
b = B()  # Create an instance of B
b.hello()  # Call inherited public method
```

```python
b.my_method()
```

```python
b.__secret()
```

```python
class Robot:
    def move(self, from_pos, to_pos):
        print(f'Moving from {from_pos} to {to_pos}')
```

```python
class FlyingRobot(Robot):
    # Override the move method
    def move(self, from_pos, to_pos):
        print(f'Flying from {from_pos} to {to_pos}')
```

```python
class SwimmingRobot(Robot):
    # Override the move method
    def move(self, from_pos, to_pos):
        print(f'Swimming from {from_pos} to {to_pos}')
```

```python
a = Robot()
b = FlyingRobot()
c = SwimmingRobot()

robots = [a, b, c]
for r in robots:
    r.move('A', 'B')  # Calls the appropriate move method based on the object type
```
